```python
#Graficar seno y coseno

x=np.linspace(0,6*np.pi,100)
y_sen = np.sin(x)
y_cos = np.cos(x)

plt.plot(x,y_sen,color="green")
plt.plot(x,y_cos,color="red")

plt.title("Gráfica del seno y coseno")
plt.xlabel("x")
plt.ylabel("y")

plt.grid(color="pink",linewidth="5",ms=20)
```

```python
x=np.linspace(0,6*np.pi,100)


def grafica(x):
  y_sen = np.sin(x)
  y_cos = np.cos(x)
  plt.plot(x,y_sen,color="green")
  plt.plot(x,y_cos,color="red")
  plt.title("Gráfica del seno y coseno")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.grid(color="pink",linewidth="5")
  plt.show()
  return None

grafica(x)
```
