```python

x=np.linspace(0,6*np.pi,100)
y_sen = np.sin(x)
plt.subplot(1,2,1)
plt.plot(x,y_sen,color="green")
plt.title("seno")
plt.grid(color="pink",linewidth="5",ms=20)

x=np.linspace(0,6*np.pi,100)
y_cos = np.cos(x)
plt.subplot(1,2,2)
plt.plot(x,y_cos,color="red")
plt.title("coseno")
plt.grid(color="pink",linewidth="5",ms=20)

plt.suptitle("Gr√°fica del seno y coseno")
plt.show()
```

```python
x=np.linspace(0,6*np.pi,100)


def grafica(x):
  y_sen = np.sin(x)
  plt.subplot(1,2,1)
  plt.plot(x,y_sen,color="green")
  plt.title("seno")
  plt.grid(color="pink",linewidth="5",ms=20)
  x=np.linspace(0,6*np.pi,100)
  y_cos = np.cos(x)
  plt.subplot(1,2,2)
  plt.plot(x,y_cos,color="red")
  plt.title("coseno")
  plt.grid(color="pink",linewidth="5",ms=20)
  plt.suptitle("Gr√°fica del seno y coseno")
  plt.show()
  return None

grafica(x)
```
