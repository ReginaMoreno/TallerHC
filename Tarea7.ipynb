```python
#promedio
arreglo = list(range(151))
pro = sum(arreglo)/len(arreglo)
print("El promedio es:",pro)
```

```python
#varianza
arreglo = list(range(151))
#calcular la media
media = sum(arreglo)/len(arreglo)
#calcular la suma de los cuadrados
difc = sum((x-media)**2 for x in arreglo)
varianza = difc / len(arreglo)
print("La varianza es:",varianza)
```

```python
#Desviación estándar
DE = varianza**.5
# Imprimir el resultado
print("La Desviación Estándar es:",DE)
```

```python
#Crear un histograma.
plt.hist(arreglo, edgecolor="pink")
plt.title("Histograma.")
plt.xlabel("Valor")
plt.ylabel("Promedio")
plt.show()
```

```python
#suma de pares e impares.
par = sum(x for x in arreglo if x % 2 == 0)
impar = sum(x for x in arreglo if x % 2 != 0)
print("La suma de los números pares en el arreglo es: ",par)
print("La suma de los números impares en el arreglo es: ", impar)
```

```python
#Múltiplos de 2, 3 y 5.
mult2 = [x for x in arreglo if x % 2 == 0]
mult3 = [x for x in arreglo if x % 3 == 0]
mult5 = [x for x in arreglo if x % 5 == 0]
print("Los múltiplos de 2 en el arreglo son:", mult2)
print("Los múltiplos de 3 en el arreglo son:", mult3)
print("Los múltiplos de 5 en el arreglo son:", mult5)
```

```python
#grafique en python la familia de curvas para la solución de la ecuación diferencial y en un sólo gráfico muestre de diferente color las curvas para los siguientes tres casos:
#para valores de C > 0
#cuando C = 0
#para valores de C < 0

#Definimos la función
def y(x, C):
    return (x - 1) + C * np.exp(-x)

#Generar valores de x
x_vals = np.linspace(8,34,89)
# Graficar para valores de C > 0
plt.plot(x_vals, y(x_vals, 2), label='C > 0', color="red")
plt.title("Valores mayores que 0")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

```

```python

x_vals = np.linspace(2, 45, 67)
# C = 0
plt.plot(x_vals, y(x_vals, 0), label='C = 0', color="cyan")
# Configurar la gráfica
plt.title("Valores igual a 0")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```

```python
x_vals = np.linspace(-3,56,78)
# C < 0
plt.plot(x_vals, y(x_vals, 0), label='C < 0', color="green")
# Configurar la gráfica
plt.title("Valores menores a 0")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```

```python
#Grafique la función con línea continua para un rango de -10 a 10, con más de 1000 muestras.

# Definir la función
def f(x):
    return (3*x + 7) / (x**2 - 4)

# Generar puntos para la gráfica
x = np.linspace(-10, 10, 1000)
y = f(x)

# Graficar la función
plt.plot(x, y, label='(3x + 7) / (x^2 - 4)')

plt.title('Gráfica de la función.')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Mostrar la gráfica
plt.show()
```

```python
plt.axhline(-1, color='green', linestyle='--', linewidth=3, label='Asíntota Horizontal')
plt.axvline(2, color='purple', linestyle='--', linewidth=3, label='Asíntota Vertical')
plt.axvline(-2, color='pink', linestyle='--', linewidth=3)

# Configuración de la gráfica
plt.title('Gráfica de la función con Asíntotas')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Mostrar la gráfica
plt.show()
```
